```python
headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'
    }
r = requests.get(url, headers=headers)
soup=BeautifulSoup(r.text,'lxml')
product=[]
price=[]
description=[]
rating=[]
for i in range(2,380):
    url='https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+str(i)
   
    box=soup.find_all('div',attrs={'data-id': True})
    
    for i in box:
        name=i.find('div',class_='RG5Slk')
        product.append(name.text)
    
    for i in box:
        pric=i.find('div',class_='hZ3P6w DeU9vF')
        price.append(pric.text)
    
    for i in box:
        des=i.find('ul',class_='HwRTzP')
        description.append(des.text)
    
    for i in box:
        rat=i.find('div',class_='MKiFS6')
        rating.append(rat.text if rat else '0')
       
df=pd.DataFrame({'ProductName':product,'prices':price,'description':description,'rating':rating})
    
```
